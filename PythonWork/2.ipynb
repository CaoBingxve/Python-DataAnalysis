折线图

```python
import pandas as pd
import matplotlib.pyplot as plt
catering_dish_profit='data/catering_sale_all.xls'
data=pd.read_excel(catering_dish_profit)
plt.rcParams['font.sans-serif']='SimHei'#用来显示正常的中文标签

plt.figure(figsize=(16,8))
plt.plot(data['日期'],data['百合酱蒸凤爪'],color='red',label='百合酱蒸凤爪',marker='o')
plt.plot(data['日期'],data['金银蒜汁蒸排骨'],color='orange',label='金银蒜汁蒸排骨',marker='s')
plt.plot(data['日期'],data['翡翠蒸香茜饺'],color='pink',label='翡翠蒸香茜饺',marker='x')
plt.plot(data['日期'],data['乐膳真味鸡'],color='yellow',label='乐膳真味鸡',marker='d')
plt.plot(data['日期'],data['蜜汁焗餐包'],color='green',label='蜜汁焗餐包',marker='D')
plt.plot(data['日期'],data['生炒菜心'],color='blue',label='生炒菜心',marker='<')
plt.plot(data['日期'],data['铁板酸菜豆腐'],color='purple',label='铁板酸菜豆腐',marker='1')
plt.plot(data['日期'],data['香煎韭菜饺'],color='white',label='香煎韭菜饺',marker='2')
plt.plot(data['日期'],data['香煎罗卜糕'],color='black',label='香煎罗卜糕',marker='p')
plt.plot(data['日期'],data['原汁原味菜心'],color='grey',label='原汁原味菜心',marker='P')
plt.legend()#显示图例
plt.ylabel('销量')
plt.show()
```
